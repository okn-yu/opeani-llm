* ada2の出力は1536次元のベクトル
* 8191トークンを入力として受け取れる
* そもそもドラゴンボールは分割する必要はないのでは...

```python
import openai

def get_embedding(text, model="text-embedding-ada-002"):
   text = text.replace("\n", " ")
   return openai.Embedding.create(input = [text], model=model)['data'][0]['embedding']
```

```python
emb = get_embedding('初音ミク')
```
